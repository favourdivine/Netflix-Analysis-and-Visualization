```python
import pandas as pd
import seaborn as sns
import numpy as np
import plotly.express as px
```

```python
df=pd.read_csv("imdb_movies_shows.csv")
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df2=df.copy()
df2["age_certification"].fillna(df["age_certification"].mode()[0], inplace=True)
```

```python
## ways you fillna to fill a datasets
## df.fillna(value)
## df.fillna(df.mean())
## df.fillna(df.median())
## df.fillna(df.mode().iloc[0])
## df.ffill()  forward fill
## df.bfill()  backward fill
## df.interpolate():Interpolate missing values based on the existing values in the column.
```

```python
df1=df.fillna(df.mode().iloc[0])
df1
```

```python
df1.isnull().sum()
```

```python
## data is cleaned now 
```

```python
df["genres"].unique()
```

```python
total_sales=df1.groupby('genres')['imdb_score'].sum()
sort_movies=total_sales.sort_values(ascending=False)
sort_movies
```

```python
total_years=df1.groupby("title")["release_year"].mean()
sort_years=total_years.sort_values(ascending=False)
sort_years
```

```python
fig=px.histogram(df1,x="type",color="age_certification",text_auto=True)
fig.show()
```

```python
fig=px.histogram(df1,x="type",y="imdb_score",color="type",text_auto=True)
fig.show()
```

```python
fig=px.histogram(df1,y="title",x="release_year",color="type",text_auto=True)
fig.show()
```

```python
## Encoding Categorical Data:
## Label Encoding: Label encoding assigns a unique numerical label to each category. 
### It's suitable when there's an inherent ordinal relationship between categories.
from sklearn.preprocessing import LabelEncoder
data = {'Category': ['Red', 'Green', 'Blue', 'Red', 'Green']}
df = pd.DataFrame(data)
label_encoder = LabelEncoder()
df['Category_Encoded'] = label_encoder.fit_transform(df['Category'])
print(df)
```
